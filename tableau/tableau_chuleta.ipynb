# 📊 Chuleta: Dashboard en Tableau

---

## 1. Construcción de las hojas

---

### 1.1 KPI — Total de pedidos

1. Arrastra `order_id` a **Marcas → Texto**.
2. Clic derecho → **Medida → Recuento distinto**.

---

### 1.2 KPI — Gasto típico

1. Arrastra `price` a **Marcas → Texto**.
2. Clic derecho → **Medida → Mediana**.
3. Formatear → **Moneda personalizada → R$**.

---

### 1.3 KPI — % Satisfechos

1. Crea el campo calculado `% Satisfechos`:
   ```
   IF [Review Score] >= 4 THEN 1 ELSE 0 END
   ```
2. Arrastra el campo a **Marcas**.
3. Clic derecho en la pastilla verde (continua) → **Medida → Promedio (Average)**.
4. Arrastra a **Texto** → clic derecho → **Formatear como porcentaje**.

---

### 1.4 ¿Desde dónde compran? — Gráfico de quesito

1. Arrastra `customer_state` a **Color** (en Marcas).
2. Arrastra `order_id` a **Marcas** → **Medida → Recuento distinto**.
3. Arrastra `order_id` a **Ángulo** (tipo circular en Marcas).
4. Arrastra `customer_state` a **Etiqueta**.
   - Para regiones poco relevantes: clic derecho en su nombre → **Etiqueta de marcas → No mostrar nunca** (así solo resaltan las 3 regiones más grandes).

> ⚠️ **Nota:** se descartó el mapa porque Tableau no reconoce los estados brasileños.  
> 💡 **Nota:** ordenar de forma descendente para mejorar la visibilidad.

---

### 1.5 ¿Qué compran? — Barras

1. Arrastra `product_category_name` a **Filas**.
2. Arrastra `order_id` a **Columnas** → **Añadir todos los miembros → Recuento distinto**.
3. Ordenar **descendente**.
4. Filtrar: `product_category_name` → **Filtros → Por campo → Top 3**.

> 💡 **Nota:** ocultar el eje → clic derecho → desmarcar **Mostrar encabezado**.

---

### 1.6 ¿Cuándo compran? — Barras

1. Arrastra `order_purchase_timestamp` a **Columnas** → clic derecho → **Día de la semana (discreto)**.
2. Arrastra `order_id` a **Filas** → **Añadir todos los miembros → Recuento**.

> 💡 **Nota:** ocultar el eje → clic derecho → desmarcar **Mostrar encabezado**.

---

### 1.7 Evolución del negocio — Líneas

1. Arrastra `order_purchase_timestamp` a **Columnas** → clic derecho → **AÑO (discreto)**.
2. Arrastra `order_id` a **Filas** → **Añadir todos los miembros → Recuento distinto**.

---

## 2. Decisiones de diseño

---

### 2.1 Filtro
Usar la opción de **embudo** para filtrar por regiones.

---

### 2.2 Imagen
Guardar una imagen → **Objetos → Insertar imagen** (recomendación: centrar).

---

### 2.3 KPIs
Colocar todos juntos en la **parte superior del dashboard**.

---

### 2.4 Formato
- **Sombreado:** Hoja de trabajo → color azul (uniforme en todas las hojas).
- **Tipografía:** Tableau Bold como estilo de letra.
  - Valores: tamaño **20**.
  - Títulos: tamaño **16** (más pequeños que los valores).

---

### 2.5 Color
- **Azul:** transmite confianza (usado por PayPal y Amazon por la misma razón).
- **Naranja-amarillo:** descartado — es el color de Mercado Livre, competidor directo de Olist.

---

### 2.6 Metodología de maquetación
- Estructura diseñada en **modo flotante**, convertida a **mosaico** para mayor estabilidad.
- Tamaño **automático** durante el desarrollo; **fijo** al publicar para evitar desajustes en Tableau Public.

---

### 2.7 Dashboard único
Un solo vistazo para contar la historia completa:
- **Arriba:** KPIs.
- **Abajo:** gráficos temporales y de detalle.
